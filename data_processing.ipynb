Importing Libraries

```python
import os
import pandas as pd
import numpy as np
```

## Combining datasets into a single dataframe

```python
# Setting Data Folder Path
```

```python
male_data = 'E:/Coding/Suicide Rate Analysis/Mortality Data/Male'
female_data = 'E:/Coding/Suicide Rate Analysis/Mortality Data/Female'
unknown_data = 'E:/Coding/Suicide Rate Analysis/Mortality Data/Unknown'
```

```python
# Creating empty dataframe
```

```python
combined_data = pd.DataFrame()
```

```python
# Listing all csv files of respective folders
```

```python
male_csv = [file for file in os.listdir(male_data) if file.endswith('.csv')]
female_csv = [file for file in os.listdir(female_data) if file.endswith('.csv')]
unknown_csv = [file for file in os.listdir(unknown_data) if file.endswith('.csv')]
```

```python
# Looping through each file and concatenating
```

```python
for csv_file in male_csv:
    file_path = os.path.join(male_data, csv_file)
    df = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, df], ignore_index=True)
combined_data
```

```python
for csv_file in female_csv:
    file_path = os.path.join(female_data, csv_file)
    df = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, df], ignore_index=True)
combined_data
```

```python
for csv_file in unknown_csv:
    file_path = os.path.join(unknown_data, csv_file)
    df = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, df], ignore_index=True)
combined_data
```

## Pre-processing Data

```python
# Dropping unecessary columns
```

```python
combined_data = combined_data.drop(['Region Code', 'Country Code', 'Age group code','Age-standardized death rate per 100 000 standard population'], axis=1)
combined_data
```

```python
# sorting data
```

```python
order = ['Region Name', 'Country Name', 'Year', 'Sex']
```

```python
df = combined_data.sort_values(by=order)
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
# Checking different age groups present in the dataframe
```

```python
df['Age Group'].unique()
```

```python
# Creating different age group columns
```

```python
columns = ['0 Year', '1-4 Years', '5-9 Years', '10-14 Years', '15-19 Years', '20-24 Years', '25-29 Years', '30-34 Years', '35-39 Years',
           '40-44 Years', '45-49 Years', '50-54 Years', '55-59 Years', '60-64 Years', '65-69 Years', '70-74 Years', '75-79 Years',
           '80-84 Years', '85+ Years', 'Unknown Age']
```

```python
df = df.assign(**{column: pd.Series(dtype='int') for column in columns})
df
```

```python
# Rearranging columns
```

```python
df = pd.concat([df.iloc[:, :5], df.iloc[:, 8:], df.iloc[:, 5:8]], axis=1)
df
```

```python
# Handling NaN values of number column
```

```python
df['Number'].fillna(0, inplace=True)
df
```

```python
# Restructuring data values
```

```python
ages = ['0', '1-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69',
        '70-74', '75-79', '80-84', '85+', 'Unknown']
```

```python
for age in ages:
    if age == '0':
        df.loc[df["Age Group"]== f"[{age}]", f"{age} Year"] = df["Number"]
    elif age == 'Unknown':
        df.loc[df["Age Group"]== f"[{age}]", f"{age} Age"] = df["Number"]
    else:
        df.loc[df["Age Group"]== f"[{age}]", f"{age} Years"] = df["Number"]

df
```

```python
# Dropping age group column
```

```python
df = df.drop(['Age Group','Number'], axis = 1)
df
```

```python
# Grouping data
```

```python
df = df.groupby(['Region Name','Country Name','Year', 'Sex'])
df = df.aggregate(np.mean)
df.reset_index(inplace=True)
df
```

```python
# Calculating number of suicide for each row
```

```python
df.insert(df.columns.get_loc('Unknown Age') + 1, 'No of Suicides', None)
df
```

```python
suicides = ['0 Year', '1-4 Years', '5-9 Years', '10-14 Years', '15-19 Years', '20-24 Years', '25-29 Years', '30-34 Years', '35-39 Years', '40-44 Years',
            '45-49 Years', '50-54 Years', '55-59 Years', '60-64 Years', '65-69 Years', '70-74 Years', '75-79 Years', '80-84 Years', '85+ Years', 'Unknown Age']
```

```python
df['No of Suicides'] = df[suicides].sum(axis=1)
df = df.fillna(0)
df
```

```python
# Eporting data
df.to_csv('who_suicide_data.csv', index=False)
```

# EDA

```python
import matplotlib.pyplot as plt
```

## Region-wise average death rate from suicide per 100000 population

```python
regional_suicides = df.groupby('Region Name')['Death rate per 100 000 population'].mean().reset_index()
regional_suicides = regional_suicides.sort_values(by='Death rate per 100 000 population', ascending=True)
```

```python
# Plotting a Pie chart
```

```python
plt.figure(figsize=(8, 8))
plt.pie(regional_suicides['Death rate per 100 000 population'], labels=regional_suicides['Region Name'], autopct='%1.1f%%', startangle=90)
plt.title('Region-wise distribution of average death from suicide per 100000 population')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## Top 10 Most Suicide prone countries

```python
country_suicides = df.groupby('Country Name')['Death rate per 100 000 population'].mean().reset_index()
country_suicides = country_suicides.sort_values(by='Death rate per 100 000 population', ascending=False)
top_10_countries = country_suicides.head(10)
```

```python
# Plotting bar graph
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries['Death rate per 100 000 population'],color='red')
plt.title('Top 10 Countries with the Most Suicide Deaths per 100 000 population')
plt.xlabel('Country Name')
plt.ylabel('Total Number of Suicides')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Suicides & Age Group

```python
age_groups = ['0 Year', '1-4 Years', '5-9 Years', '10-14 Years', '15-19 Years',
              '20-24 Years', '25-29 Years', '30-34 Years', '35-39 Years', '40-44 Years',
              '45-49 Years', '50-54 Years', '55-59 Years', '60-64 Years', '65-69 Years',
              '70-74 Years', '75-79 Years', '80-84 Years', '85+ Years', 'Unknown Age']
total_suicides_by_age_group = df[age_groups].sum()
```

```python
# Plotting Histogram
```

```python

plt.figure(figsize=(12, 6))
plt.bar(total_suicides_by_age_group.index, total_suicides_by_age_group.values)

plt.title('Distribution of Suicide Deaths Across Age Groups (All Years)')
plt.xlabel('Age Group')
plt.ylabel('Total Number of Suicides')
plt.xticks(rotation=90)

# Disable scientific notation on the y-axis
plt.ticklabel_format(style='plain', axis='y')

plt.grid(axis='y')
plt.show()
```

## Distribution of suicide numbers by gender

```python
total_suicides_by_gender = df.groupby('Sex')['Death rate per 100 000 population'].mean()
```

```python
# Plotting pie chart
```

```python
plt.figure(figsize=(8, 8))
plt.pie(total_suicides_by_gender, labels=total_suicides_by_gender.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender-wise Distribution of death from suicides per 100,000 Population')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
suicides_by_year_gender = df.groupby(['Year', 'Sex'])['No of Suicides'].sum().unstack()
```

```python
# Plotting line graph
```

```python
plt.figure(figsize=(12, 6))
plt.plot(suicides_by_year_gender.index, suicides_by_year_gender['Male'], label='Male', marker='o')
plt.plot(suicides_by_year_gender.index, suicides_by_year_gender['Female'], label='Female', marker='o')
plt.title('Number of death from suicides over the years')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.legend()
plt.grid(True)
plt.show()
```
