```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("who_suicide_statistics.csv")
```

```python
df.loc[df["sex"]=='male', "Male"] = df["suicides_no"]
```

```python
df.loc[df["sex"]=='female', "Female"] = df["suicides_no"]
```

```python
df.loc[df["age"]=="5-14 years", "5-14"] = df["suicides_no"]
```

```python
df.loc[df["age"]=="15-24 years", "15-24"] = df["suicides_no"]
```

```python
df.loc[df["age"]=="25-34 years", "25-34"] = df["suicides_no"]
```

```python
df.loc[df["age"]=="35-54 years", "35-54"] = df["suicides_no"]
```

```python
df.loc[df["age"]=="55-74 years", "55-74"] = df["suicides_no"]
```

```python
df.loc[df["age"]=="75+ years", "75+"] = df["suicides_no"]
```

```python
df=df.drop(['age'], axis=1)
```

```python
df=df.dropna(subset=['suicides_no'])
```

```python
group1 = df.groupby(['country', 'year','sex'])
group1 = group1.aggregate(np.sum)
```

```python
group1.to_csv('file01.csv')
```
